# StackOverflow Train Stats via scripts/report

This notebook runs report scripts from this repo on your dataset folder.

Dataset folder used here: `/tempory/the_three_potatoes/ri_project/workspaces/output/stackoverflow/train`.

```python
from pathlib import Path
import os
import subprocess

REPO_ROOT = Path('/tempory/the_three_potatoes/ri_project/workspaces/amelie/FlexNeuART-IR-TTP')
DATA_DIR = Path('/tempory/the_three_potatoes/ri_project/workspaces/output/stackoverflow/train') # change to other dataset 
TMP_DIR = REPO_ROOT / 'scripts' / 'report' / 'tmp'
TMP_DIR.mkdir(parents=True, exist_ok=True)

print('REPO_ROOT =', REPO_ROOT)
print('DATA_DIR  =', DATA_DIR)
print('TMP_DIR   =', TMP_DIR)
```

## 1) Prepare temporary JSONL inputs (needed by print_field_tok_stat.py)
`scripts/report/print_field_tok_stat.py` expects `--input` as JSONL and `--field` name.
Your `question_text` / `answer_text` files are plain one-line-per-record text, so we wrap each line into JSONL records with field `text`.

```python
cmd = r'''
set -e
cd /tempory/the_three_potatoes/ri_project/workspaces/output/stackoverflow/train

python - << 'PY'
import json
from pathlib import Path

data_dir = Path('/tempory/the_three_potatoes/ri_project/workspaces/output/stackoverflow/train')
tmp_dir = Path('/tempory/the_three_potatoes/ri_project/workspaces/amelie/FlexNeuART-IR-TTP/scripts/report/tmp')
tmp_dir.mkdir(parents=True, exist_ok=True)

pairs = [
    ('question_text', 'question_text.jsonl'),
    ('answer_text', 'answer_text.jsonl'),
    ('question_bigram', 'question_bigram.jsonl'),
    ('answer_bigram', 'answer_bigram.jsonl')
]

for src_name, dst_name in pairs:
    src = data_dir / src_name
    dst = tmp_dir / dst_name
    with src.open('r', encoding='utf-8', errors='ignore') as fi, dst.open('w', encoding='utf-8') as fo:
        for i, line in enumerate(fi):
            rec = {'DOCNO': str(i), 'text': line.rstrip('\n')}
            fo.write(json.dumps(rec, ensure_ascii=True) + '\n')
    print('created', dst)
PY
'''

subprocess.run(cmd, shell=True, check=True, text=True)
```

## 2) Run scripts/report/print_field_tok_stat.py
These are direct calls to a script in `scripts/report`.

```python
cmd = r'''
set -e
cd /tempory/the_three_potatoes/ri_project/workspaces/amelie/FlexNeuART-IR-TTP

python scripts/report/print_field_tok_stat.py --input /tempory/the_three_potatoes/ri_project/workspaces/amelie/FlexNeuART-IR-TTP/scripts/report/tmp/question_text.jsonl --field text
python scripts/report/print_field_tok_stat.py --input /tempory/the_three_potatoes/ri_project/workspaces/amelie/FlexNeuART-IR-TTP/scripts/report/tmp/answer_text.jsonl --field text
python scripts/report/print_field_tok_stat.py --input /tempory/the_three_potatoes/ri_project/workspaces/amelie/FlexNeuART-IR-TTP/scripts/report/tmp/question_bigram.jsonl --field text
python scripts/report/print_field_tok_stat.py --input /tempory/the_three_potatoes/ri_project/workspaces/amelie/FlexNeuART-IR-TTP/scripts/report/tmp/answer_bigram.jsonl --field text
'''

subprocess.run(cmd, shell=True, check=True, text=True)
```

##### For other dataset than train (with qrels summary)

```python
import json
import shutil
from pathlib import Path

def clean_tmp_dir():
    if TMP_DIR.exists():
        shutil.rmtree(TMP_DIR)
    TMP_DIR.mkdir(parents=True, exist_ok=True)

def build_tmp_jsonl_for_split(split_dir: Path):
    pairs = [
        ('question_text', 'question_text.jsonl'),
        ('answer_text', 'answer_text.jsonl'),
        ('question_bigram', 'question_bigram.jsonl'),
        ('answer_bigram', 'answer_bigram.jsonl')
    ]

    for src_name, dst_name in pairs:
        src = split_dir / src_name
        if not src.exists():
            raise FileNotFoundError(f'Missing input file: {src}')

        dst = TMP_DIR / dst_name
        with src.open('r', encoding='utf-8', errors='ignore') as fi, dst.open('w', encoding='utf-8') as fo:
            for i, line in enumerate(fi):
                rec = {'DOCNO': str(i), 'text': line.rstrip('\n')}
                fo.write(json.dumps(rec, ensure_ascii=True) + '\n')

def run_report_stats_from_tmp(split_name: str):
    cmd_final = f'''
set -e
cd {REPO_ROOT}

python scripts/report/print_field_tok_stat.py --input {TMP_DIR}/question_text.jsonl --field text
python scripts/report/print_field_tok_stat.py --input {TMP_DIR}/answer_text.jsonl --field text
python scripts/report/print_field_tok_stat.py --input {TMP_DIR}/question_bigram.jsonl --field text
python scripts/report/print_field_tok_stat.py --input {TMP_DIR}/answer_bigram.jsonl --field text
'''

    print(f'\n=== {split_name} : token stats ===')
    subprocess.run(cmd_final, shell=True, check=True, text=True)

def run_qrels_summary(split_dir: Path, split_name: str):
    cmd = f'''
set -e
cd {split_dir}

echo 'line counts:'
wc -l question_text answer_text question_bigram answer_bigram qrels.txt

echo
echo 'qrels summary:'
awk '{{pairs++; q[$1]=1; d[$3]=1; g[$4]++}} END{{
  print "pairs="pairs, "queries="length(q), "docs="length(d);
  for (k in g) print "grade", k, "count", g[k];
}}' qrels.txt | sort -k2,2n
'''

    print(f'\n=== {split_name} : qrels stats ===')
    subprocess.run(cmd, shell=True, check=True, text=True)
```

```python
# Add/remove split names here
split_names = ['test', 'dev1', 'dev2', 'tran']

for split_name in split_names:
    split_dir = DATA_DIR.parent / split_name
    print(f'\n----- Processing split: {split_name} ({split_dir}) -----')

    if not split_dir.exists():
        print(f'Skipping: folder does not exist -> {split_dir}')
        continue

    try:
        clean_tmp_dir()
        build_tmp_jsonl_for_split(split_dir)
        run_report_stats_from_tmp(split_name)
        run_qrels_summary(split_dir, split_name)
    except Exception as e:
        print(f'Error in split {split_name}: {e}')
```
