### Reliable MGIZA build workaround (CMake 4.x compatible)
If `flexneuart_install_extra.sh ... 1` fails with a CMake policy/minimum error, run the commands below.

They build MGIZA directly in `./flexneuart_scripts/mgiza/mgizapp/bin`, which is what FlexNeuART scripts use.

```python
!cd ./flexneuart_scripts && \
  rm -rf mgiza && \
  git clone https://github.com/moses-smt/mgiza.git && \
  cd mgiza/mgizapp && \
  cmake -DCMAKE_POLICY_VERSION_MINIMUM=3.5 . && \
  make -j4 && \
  ls -la bin
```
